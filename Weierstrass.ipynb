```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def weierstrass(x,a,b,n):
    n_series=np.arange(0,n,1)
    term1=a**n_series
    term2=np.cos(b**n_series*np.pi*x)
    term=term1*term2
    return np.sum(term)
n_max=10000
n_step=1000
a=0.5
b=20
check= a*b>1+1.5*np.pi
print(check)
x=np.linspace(-2,2,n_step)
vfunc=np.vectorize(weierstrass)

y=vfunc(x,a,b,n_max)

plt.plot(x,y)
plt.show()
```
